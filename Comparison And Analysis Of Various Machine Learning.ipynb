# Feature Selection by finding Covariance :

Covariance tells how one variable is dependent on other. It helps to do features selection.

We have to make sure that one variable is not heavily dependent on other variable.

```python
df.corr()
```

This gives us just the covariance numbers. We can visualize it through seaborn.

```python
sb.heatmap(df.corr())
```

```python
#To make that map bigger
from matplotlib.pyplot import rcParams
rcParams['figure.figsize']=14,12
```

```python
sb.heatmap(df.corr())
```

This Heatmap will help you to visualize with help of colors.

" The lighter the color in the map --> represents positive covariance "

" The darker the color in the map --> represents negative covariance "

```python
#To represent values of covariance on the map.
sb.heatmap(df.corr(),annot=True)
```

If the covariance is close to zero(positive or negative), then features are independent of each other.


If the value is more positive(close to 1), then features are positively codependent and If the value is more negative(close to -1) then features are negatively codependent.

Positively codependent ->If one value increases then other value will also increase || Negatively codependent ->If one value increases then other value will decrease.


If we have high positive covariance or high negative covariance among 2 features, then we should neglect one of the feature while training the model. But in the above features we dont find huge positive or negative covariance. so we can consider all features(columns) for model construction and training.

# 4. ADABOOST CLASSIFIER :

.

<img src='maxresdefault.jpg'/>

An AdaBoost classifier is a meta-estimator that begins by fitting a classifier on the original dataset and then fits additional copies of the classifier on the same dataset but where the weights of incorrectly classified instances are adjusted such that subsequent classifiers focus more on difficult cases.

.

```python
from sklearn.model_selection import train_test_split
X=df.drop('output',axis=1)
Y=df[["output"]]
x_train, x_test, y_train, y_test =train_test_split(X, Y, test_size = 0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy', max_depth=2)
from sklearn.ensemble import AdaBoostClassifier
# Strong learner
AdaBoost = AdaBoostClassifier(base_estimator=model, n_estimators=400)
```

```python
#Fit the model with training data
boostmodel = AdaBoost.fit(x_train.values,y_train.values.ravel())
boostmodel
```

```python
#Evaluate the accuracy of the model
from sklearn import metrics
y_pred = boostmodel.predict(x_test.values)
print(y_pred)
metrics.accuracy_score(y_test, y_pred)
#Calculating the accuracy in percentage
```

<b> "So, accuracy is 81.01 %, if we use Adaboost classifier". </b>

# 5. RANDOM FOREST CLASSIFIER :

<img src='random forest.jpg'/>

The random forest is a classification algorithm consisting of many decisions trees. It uses bagging and feature randomness when building each individual tree to try to create an uncorrelated forest of trees whose prediction by committee is more accurate than that of any individual tree.

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train.values.ravel())
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds=roc_curve(y_test, y_pred)
roc_auc= auc(false_positive_rate, true_positive_rate)
roc_auc
```

```python
n_estimators=[1,2,4,8,16,32,64,128,256,512]
train_results=[]
test_results=[]

for estimator in n_estimators:
    rf=RandomForestClassifier(n_estimators=estimator, n_jobs=-1)
    rf.fit(x_train,y_train.values.ravel())
    train_pred=rf.predict(x_train)
    false_positive_rate, true_positive_rate, thresholds=roc_curve(y_train, train_pred)
    roc_auc=auc(false_positive_rate, true_positive_rate)
    train_results.append(roc_auc)
    y_pred=rf.predict(x_test)
    false_positive_rate, true_positive_rate, thresholds=roc_curve(y_test, y_pred)
    roc_auc=auc(false_positive_rate, true_positive_rate)
    test_results.append(roc_auc)

from matplotlib.legend_handler import HandlerLine2D
line1, =plt.plot(n_estimators, train_results,'b', label='Train AUC')
line2, =plt.plot(n_estimators, test_results,'r', label='Test AUC')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.show()
```
